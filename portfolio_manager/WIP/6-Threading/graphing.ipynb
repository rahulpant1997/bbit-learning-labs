```python
import os
import sys
module_path = os.path.abspath('..')
if module_path not in sys.path:
    sys.path.append(module_path)
from implementations.securitySolution import security
from datetime import datetime
from time import sleep
import numpy as np

def create_security_mv(securityName, dataPointSize):
    size = dataPointSize
    securityObj = security(securityName)
    MV = {}
    count = 0
    while count < size:
        MV[count] = securityObj.get_current_market_value()
        count += 1

    x_data = list(MV.keys())
    y_data = list(MV.values())

    plt.figure(title=securityObj.get_name(), animation_duration=1000)
    plt.Lines.
    plt.plot(x_data, y_data)
    plt.show()
    return plt
    
create_security_mv("IBM US Equity", 1000)
```
